```python
import json
import math
import numpy as np
import pandas as pd
import datetime
```

```python
data = pd.read_excel('../data/raw/raw_data.xlsx')
```

```python
data_small = data[['id', 'responsibilities(Должностные обязанности)', 'requirements(Требования к соискателю)', 'terms(Условия)', 'notes(Примечания)']]
```

```python
data_small = data_small.fillna(value='')
```

```python
data_small.head()
```

```python
data_small.loc[data_small[data_small['notes(Примечания)'] == '\\N'].index, 'notes(Примечания)'] = ''
```

```python
data_small.dtypes
```

```python
def parse_indexes(big_sent, small_sent):
    if len(small_sent) != 0 and small_sent in big_sent:
        start_ind = big_sent.index(small_sent)
        return (start_ind, start_ind + len(small_sent))
    return -1
```

```python
data_small.head()
```

```python
data_small.loc[data_small[data_small['id'] == 9492540].index, 'terms(Условия)'] = '5/2'
```

```python
for n, col in enumerate(data_small.columns[1:]):
    data_small[f'{col}_extracted_indexes'] = data_small.apply(lambda row: parse_indexes(row['responsibilities(Должностные обязанности)'], row[col]), axis=1)
```

```python
data_small.head()
```

```python
data_json = []
```

```python
for index, row in data_small.iterrows():
    if row['requirements(Требования к соискателю)_extracted_indexes'] != -1:
        req_start, req_end = row['requirements(Требования к соискателю)_extracted_indexes']
    else:
        req_start, req_end = -1, -1

    if row['terms(Условия)_extracted_indexes'] != -1:
        ter_start, ter_end = row['terms(Условия)_extracted_indexes'] 
    else:
        ter_start, ter_end = -1, -1

    if row['notes(Примечания)_extracted_indexes'] != -1:
        notes_start, notes_end = row['notes(Примечания)_extracted_indexes']
    else:
        notes_start, notes_end = -1, -1
    
    data_json.append({
        'id': row['id'],
        'text': row['responsibilities(Должностные обязанности)'],
        'extracted_part': {
            'requirements': {
                'start_index': req_start,
                'end_index': req_end,
                'text' : row['requirements(Требования к соискателю)'],
            },
            'terms': {
                'start_index': ter_start,
                'end_index': ter_end,
                'text': row['terms(Условия)'],
            },
            'notes': {
                'start_index': notes_start,
                'end_index': notes_end,
                'text': row['notes(Примечания)'],
            }
        }
    })
```

```python
with open("data.json", "w") as final:
   json.dump(data_json, final, indent=4)
```
